# 定義 server function

```python
from clip_diffusion.utils.anvil_utils import *
```

```python
from clip_diffusion.sample import guided_diffusion_sample, latent_diffusion_sample
```

```python
DIFFUSION_MODEL_PATHS = {
    "通用": None,
    "景觀": f"{MODEL_PATH}/stable_diffusion_landscape_ema_0.9999_381000.pt",
    "建築": f"{MODEL_PATH}/building_ema_0.9999_170000.pt"
}
```

```python
@anvil.server.callable
def guided_sample(
    prompt,
    use_auto_modifiers,
    num_modifiers,
    model_type,
    seed,
    init_image,
    steps,
    skip_timesteps,
    eta,
    num_batches,
    display_rate
):
  print(DIFFUSION_MODEL_PATHS[model_type])
  task = anvil.server.launch_background_task(
      "guided_diffusion_sample",
      prompt=prompt,
      use_auto_modifiers=use_auto_modifiers,
      num_modifiers=num_modifiers,
      custom_model_path=DIFFUSION_MODEL_PATHS[model_type],
      seed=seed,
      init_image=init_image,
      steps=steps,
      skip_timesteps=skip_timesteps,
      eta=eta,
      num_batches=num_batches,
    )
  return task
```

```python
@anvil.server.callable
def latent_sample(
    prompt,
    seed,
    init_image,
    mask_image,
    sample_mode,
    diffusion_steps,
    eta,
    latent_diffusion_guidance_scale,
    num_iterations,
    num_batches
):
  task = anvil.server.launch_background_task(
      "latent_diffusion_sample",
      prompt=prompt,
      seed=seed,
      init_image=init_image,
      mask_image=mask_image,
      sample_mode=sample_mode,
      diffusion_steps=diffusion_steps,
      eta=eta,
      latent_diffusion_guidance_scale=latent_diffusion_guidance_scale,
      num_iterations=num_iterations,
      num_batches=num_batches
    )
  return task
```
